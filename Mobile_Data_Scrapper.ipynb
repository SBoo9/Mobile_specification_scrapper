```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.91mobiles.com/top-10-mobiles-in-india'
```

```python
page = requests.get(url, verify=False)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
collector=soup.find_all('span', class_='a2 star star_width')
```

```python
i=0
for item in collector:
    link=item.find('a', class_="hover_blue_link name gaclick")
    
    if link:
        print(link.get('href'))
        print(link.get('title'))
        #i=i+1
        #data=[link.get('href'),link.get('title')]
        #df = pd.DataFrame(data, columns=col)
```

```python
import pandas as pd
```

```python
col=['Name','ref_link']
```

```python
df=pd.DataFrame(columns=col)
```

```python
df
```

```python
i=0
for item in collector:
    link=item.find('a', class_="hover_blue_link name gaclick")
    
    if link:
        href=(link.get('href'))
        title=(link.get('title'))
        i=i+1
        data=[[title,href]]
        df = pd.concat([df, pd.DataFrame(data, columns=col)])
```

```python
pd.set_option('display.max_rows', None)
df
```

```python
for i in range(0,34):
    df.drop(index=i,inplace=True)
```
